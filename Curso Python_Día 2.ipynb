##Elif
Ahora hagamos un programa en el que preguntamos la calificación del 0 al 100 y queremos saber su equivalente en el sistema de evaluación estadounidense. Para simplicidad, basémonos en esto:
- A+ = 100
- A = 90 - 99
- B = 80 - 89
- C = 70 - 79
- D = 60 - 69
- F = < 60

```python
cal = 95
if cal == 100:
  print("A+")
elif cal >= 90:
  print("A")
elif cal >= 80:
  print("B")
elif cal >= 70:
  print("C")
elif cal >= 60:
  print("D")
else: # notemos como al ya no haber una condición, el else cubre todo lo demás
  print("F")
```

## Operadores Lógicos con Condicionales

```python
# and, or, not
pan = True
queso = True
mayonesa = True
jamón = False
print(pan and queso)
print(pan and not queso)
print(pan and not queso and mayonesa and not jamón)
```

```python
# and, or, not
pan = True
queso = True
mayonesa = True
jamón = False
print(pan and queso and mayonesa and jamón)
if not(pan and queso and mayonesa and jamón):
  print("No te puedes hacer el sandwich")
```

```python
cal = -3
if cal >= 0 and cal <= 100:
  if cal == 100:
    print("A+")
  elif cal >= 90:
    print("A")
  elif cal >= 80:
    print("B")
  elif cal >= 70:
    print("C")
  elif cal >= 60:
    print("D")
  else: # notemos como al ya no haber una condición, el else cubre todo lo demás
    print("F")
else:
  print("Error: La calificación debe estar entre 0 y 100")
```

```python
# El módulo (%)
# El módulo es un operador matemático que devuelve el residuo de una división
# 10 % 3 da 1, porque 10 dividido entre 3 cabe 3 veces y sobra 1
# 7 % 2 da 1, porque el 7 no es divisible de forma exacta entre 2
# 2 % 2 -> 0
num = 9
if num % 3 == 0 or num % 5 == 0:
  print(f"El número {num} es múltiplo de 3 O 5")

# Saber si el número es par o impar usando not
num = 5
if not num % 2 == 0:
  print(f"El número {num} es impar")
```

## Ciclo For
Los llamados ciclos for, nos ayudan a realizar acciones durante una cantidad determinada de veces

```python
# Imprimir los números del 0 al 4
for i in range(5): # Toma 5 como el número final pero inicia SIEMPRE en 0
  print(i) # Terminando el for, deja de existir la variable
```

```python
for num in range(2, 11):
  print(num)
```

```python
# También existe el range (inicio, fin, paso)
for i in range(2, 21, 2):
  print(i)
```

```python
# También existe el range (inicio, fin, paso)
n = 10
for i in range(2, n + 5, 2):
  print(i)
```

```python
cadena = "Python"
for letra_por_letra in cadena:
  print(letra_por_letra)
```

## While
Funciona siempre hasta que la condición se vuelva falsa

```python
suma =  0
números = int(input("Ingresa un número para sumar o 0 para salir"))

while números != 0: # != es equivalente a "distinto de..."
  suma += números
  números = int(input("Ingresa un número para sumar o 0 para salir"))
print(f"La suma es: {suma}")
```

```python
# Usar el while como for
número = 0
while número <= 12:
  print(número)
  número += 1 # Hay que ponerle un iterador para que no se cicle
```

```python
print("Juego: Adivina el Número")
número_secreto = int(input("Jugador 1, escribe el número secreto (1-20): "))

intentos = 0
max_intentos = 4

while intentos < max_intentos:
    intento = int(input("Adivina el número: "))
    intentos += 1

    if intento == número_secreto:
        print("Correcto")
        break
    else:
        if intentos < max_intentos:
            print("Incorrecto, intenta de nuevo")

if intentos == max_intentos and intento != número_secreto:
    print("Se acabaron los intentos")
    print("Perdiste, el número era: ", número_secreto)
```
