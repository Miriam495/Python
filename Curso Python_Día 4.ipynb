##Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adiós, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + "\n" + despedida

print(mensaje_completo("José"))
```

```python
# Ahora supongamos que tenemos una heladería
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "menta"]
  print("Sabor disponible", ",", sabores)
  # lower() hace minúsculas las palabras
  # strip() elimina los espacios tanto antes como después
  sabor = input("Elige un sabor: ").strip().lower()
  # si sabor NO está en sabores
  while sabor not in sabores:
    sabor = input("Elige un sabor nuevamente: ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("¿Cuántas bolas de nieve quieres?: "))
  while cantidad <= 0:
    cantidad = int(input("¿Cuántas bolas de nieve quieres? (inngresa un valor válido): "))
  return cantidad

# Función Madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("¿Cuántos clientes atenderás?: "))
for i in range(total_pedidos):
  pedir_helado()
```

## Funciones Lambda
Una función lambda es una función acortada para escribir una función de esta forma:

nombreFunción = lambda parámetros: acción

La diferencia entre las funciones lambda y las funciones normales es que suelen ser un poco más limitadas.

```python
sumar = lambda x,y : x+y
print(sumar(2,3))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1-desc)), (precio * desc))
nuevoPrecio = descuentoLambda(0.15, 400)
print("El precio a pagar es", nuevoPrecio[0], ", el descuento que se hizo fue de: ", nuevoPrecio[1])
print(type(nuevoPrecio))
```

```python
#Primer ejemplo: Una función que recibe una lista y obtiene el promedio del grupo y el mayor promedio
#NOTA: Solo consideramos cuál es el mayor, no cuántas veces se puede repetir ese promedio
def promediosYmayor(promedios):
  promedio = 0
  for x in promedios:
    promedio += x
  promedio = promedio / len(promedios)
  return promedio, max(promedios)

promedios = []
continuar = ""

while True:
  if len(promedios) > 0:
    continuar = ""
    while continuar not in ("si", "no"):
      continuar = input("Ingrese una respuesta válida: si/no ").lower()
    if continuar == "no":
      break

  while True:
    try:
      calif = int(input("Indique la calificación: "))
      if calif < 0 or calif > 100:
        print("Ingrese un valor válido")
      else:
        promedios.append(calif)
        break
    except ValueError:
      print("Ingrese un valor numérico")

prom, mayor = promediosYmayor(promedios)

print("\nResultados:")
print("Promedio:", prom)
print("Mayor:", mayor)
```

```python
# Calculadora -> Suma, resta, producto, división y potencia
def suma(a,b):
  return a + b
def resta(a,b):
  return a - b
def producto(a,b):
  return a * b
def división(a,b):
  if b == 0:
    return "Error: no se puede dividir entre 0"
  return a / b
def potencia(a,b):
  return a ** b

def menú():
  print("--------MI CALCULADORA--------")
  print("1. Suma")
  print("2. Resta")
  print("3. Producto")
  print("4. División")
  print("5. Potencia")
  print("6. Salir")

while True:
  menú()
  try:
    op = int(input("Seleccione una opción:"))
  except ValueError:
    print("Ingrese un número válido")
    continue

  if op == 6:
    print("SALIENDO...")
    break

  if op < 1 or op > 6:
    print("Opción inválida")
    continue

  try:
    a = float(input("Ingrese el primer número: "))
    b = float(input("Ingrese el segundo número: "))
  except ValueError:
    print("Ingrese un número válido")
    continue

  if op == 1:
    print("Resultado: ", suma(a,b))
  elif op == 2:
    print("Resultado: ", resta(a,b))
  elif op == 3:
    print("Resultado: ", producto(a,b))
  elif op == 4:
    print("Resultado: ", división(a,b))
  elif op == 5:
    print("Resultado: ", potencia(a,b))
```
